```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
c = [1,3,5,8]
a = [0.138,0.419,0.807,1.105]
```

```python

#data

data = (c,a)
data = pd.DataFrame(data)
data.index = ['c',"a"]  # type: ignore
data = data.T
data
```

```python



#plot a linear regression graph
sns.regplot(x=c,y=a,color="blue", marker="+",scatter_kws={"s": 30},line_kws={"linewidth": 0.5}, ci=None,truncate=True,order=1)

plt.xlabel("c")
plt.ylabel("a")
plt.xlim(0,10)
plt.ylim(0,1.2)
plt.title("c / a")
plt.grid(   axis="both",    which="both",    color="black",    linestyle="-",    linewidth=0.5)
plt.savefig("adsorb.png")
plt.show()





```

```python
cd = [-0.2369,0.4855,0.9005,1.1930]
qd = [1.4377,2.1252,2.5288,2.8155]
```

```python

#data

data = (cd,qd)
data = pd.DataFrame(data)
data.index = ['cd','qd']  # type: ignore
data # type: ignore


#plot a linear regression graph
sns.regplot(cd,qd, color="blue", marker="+",scatter_kws={"s": 30},    line_kws={"linewidth": 0.5}, ci=None,truncate=True,order=1,robust=True) # type: ignore
plt.xlabel("cd")
plt.ylabel("qd")

plt.title("cd / qd")
plt.grid(   axis="both",    which="both",    color="black",    linestyle="-",    linewidth=0.5)
plt.savefig("adsorb2.png")
plt.show()





```
